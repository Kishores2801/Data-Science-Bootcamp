## Import the relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Categorical.csv'</i>.

```python
data =  pd.read_csv("Categorical.csv")
data.head()
```

```python
data["continent"].unique()
```

## Map the data
Use the <i>'continent'</i> category for this analysis.

```python
data_mapped = data.copy()
```

```python
data_mapped["continent"] = data_mapped["continent"].map({'North America':0 , 'Asia': 1, 'Africa':2 ,
                                           'Europe': 3,  'South America': 4, 'Oceania': 5, 
                                           'Antarctica': 6, 'Seven seas (open ocean)':7})
```

## Select the features

```python
x = data.iloc[:, 1:3]
x.head()
```

## Clustering

Use 4 clusters initially.

```python
Kmeans = KMeans(4)
```

## Clustering results

```python
Kmeans.fit(x)
```

```python
indentified_clusters = Kmeans.fit_predict(x)
indentified_clusters
```

## Plot the data

```python
data_with_clusters = data_mapped.copy()
data_with_clusters["Cluster"] = indentified_clusters
data_with_clusters
```

```python
plt.scatter(x=data_with_clusters["Longitude"], y=data_with_clusters["Latitude"], c=data_with_clusters["Cluster"], cmap='rainbow')
plt.ylim(-90, 90)
plt.xlim(-180, 180)
plt.show()
```

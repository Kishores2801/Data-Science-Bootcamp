## Making the prediction for Real Estates

### Import the Libraries

```python
# import pandas, numpy, matplotlib, seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

### Load the Data

```python
data =  pd.read_csv('real_estate_price_size_year.csv')
data.head()
```

```python
data.describe()
```

### Multiple Linear Regression
#### Declare Dependent, Independent Variables

```python
# There are two independent variables: 'SAT' and 'Rand 1,2,3'
x = data[['size','year']]
# and a single dependent variable: 'GPA'
y = data['price']
```

### Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
x_scaled = scaler.transform(x)
```

```python
x_scaled
```

#### Muliple Linear Regression

```python
reg = LinearRegression()
reg.fit(x_scaled,y)
```

```python
reg.coef_
```

```python

reg.intercept_
```

### Summary Table

```python
reg_summary = pd.DataFrame([["Bias"], ["Size"], ["Year"]])
reg_summary
```

```python
reg_summary["Weights"] = reg.intercept_, reg.coef_[0], reg.coef_[1]
reg_summary
```
